```python
def data_loader(des_path,size):
    a=os.listdir(des_path)
    X=[]
    Y=[]
    m=0
    for i in a:
        b=des_path+'/'+str(i)
        g=os.listdir(b)
        for k in g:
          path=b+'/'+k
          img=cv2.imread(path)
          img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
          img=cv2.resize(img,size)
          #print(img.shape)
          X.append(np.array(img))
          Y.append(m)
        m+=1 
    return(X,Y)
```

```python
X,Y = data_loader(des_path='/content/contentdrive/MyDrive/GameData',size=(300,300))
```

```python
X_train_shape, X_test_shape, Y_train_shape, Y_test_shape=train_test_split(X, to_categorical(np.array(Y)),test_size=0.1,random_state=0)
```

```python
Image_size=[300,300,3]
densenet=tf.keras.applications.DenseNet121(input_shape=Image_size,weights='imagenet',include_top=False)
for layer in densenet.layers:
    layer.trainable=True
x=Flatten()(densenet.output)
prediction=Dense(2,activation='softmax')(x)
model=Model(inputs=densenet.input,outputs=prediction)
model.summary()
model.compile(loss='binary_crossentropy',optimizer='sgd',metrics=['accuracy'])
```

```python
checkpoint=tf.keras.callbacks.ModelCheckpoint('/fashion_model_23.h5',verbose=1,save_best_only=True)
callbacks=[tf.keras.callbacks.EarlyStopping(patience=2,monitor='val_loss')]
```

```python
r=model.fit(np.array(X_train_shape),Y_train_shape, batch_size=32,verbose=1,epochs=30,validation_split=0.2,shuffle=True,callbacks=callbacks)
```

```python
model.evaluate(np.array(X_test_shape),Y_test_shape)
```
